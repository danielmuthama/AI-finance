### B3

```python
import yfinance as yf
import pandas as pd

ticker = yf.Tickers('msft aapl ibm')

def get_historical_data(ticker, start_date):
    data = yf.download(ticker, start=start_date)
    #daily return
    data['Daily Return'] = data ['Adj Close'].pct_change()
    return data.dropna()
```

```python
def std_dev(data):
    #get number of observation
    n=len(data)
    #calculate mean
    mean=sum(data)/n
    #calculate deviations from mean
    deviations = sum([(x-mean)**2 for x in data])
    #calculate variance&standard deviation
    variance = deviations/(n-1)
    s = variance**(1/2)
    return s
```

```python
# calculate sharpe ratio

def sharpe_ratio(data, risk_rate=0.0):
    # Calculate average daily return
    mean_daily_return = sum(data)/len(data)
    # Calculate Standard Deviation
    s = std_dev(data)
    # Calculate Daily Sharpe Ratio
    daily_sharpe_ratio = (mean_daily_return - risk_rate)/s
    #Annualize Daily Sharpe Ratio
    sharpe_ratio = 252**(1/2)*daily_sharpe_ratio
    return sharpe_ratio

```

```python
AAPL = get_historical_data('AAPL', start_date='2022-02-03')
sharpe_ratio(AAPL['Daily Return'])
```

```python
MSFT = get_historical_data('MSFT', start_date='2022-02-03')
sharpe_ratio(MSFT['Daily Return'])
```

```python
IBM = get_historical_data('IBM', start_date='2022-02-03')
sharpe_ratio(IBM['Daily Return'])
```
